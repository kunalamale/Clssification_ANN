```python
import tensorflow as tf
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("/content/drive/MyDrive/Deep Learning/DL_Mam/Churn_Modelling.csv")
```

```python
data.head()
```

```python
#Drop the columns (RowNumber,CustomerId,Surname)
```

```python
x = data.iloc[:,3:-1]
```

```python
x
```

```python
y = data.iloc[:,-1]
```

```python
y
```

```python
x.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
```

```python
oe = OrdinalEncoder()
```

```python
x.iloc[:,[1,2]] = oe.fit_transform(x.iloc[:,[1,2]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
sc = StandardScaler()
```

```python
xtrain
```

```python
xtrain = sc.fit_transform(xtrain)
```

```python
xtest = sc.transform(xtest)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense

from sklearn.metrics import classification_report
```

```python
# 1 inisilization model
ann = Sequential()
```

```python
#2creating hiden layer
ann.add(Dense(units=6, activation= "relu"))
ann.add(Dense(units=1, activation= "sigmoid"))
```

```python
#3 establishing the connection
ann.compile(optimizer="adam",loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
#4 training model
ann.fit(xtrain,ytrain, batch_size = 30, epochs = 100)
```

```python
#5 make prediction
ypred = ann.predict(xtest)
```

```python
ypred 
```

```python
print(classification_report(ytest, ypred.round()))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(xtrain,ytrain)
pred=lg.predict(xtest)
```

```python
print(classification_report(ytest, pred))
```
